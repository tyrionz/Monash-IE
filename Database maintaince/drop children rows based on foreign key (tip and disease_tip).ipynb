```python
import pandas as pd
```

```python
tip = pd.read_excel('Disease table by team for iteration 2.5.xlsx', sheet_name = 'Tip')
```

```python
tip
```

```python
disease_tip = pd.read_excel('Disease table by team for iteration 2.5.xlsx', sheet_name = 'Disease_Tip')
```

```python
tip_tipID = tip['TIP_ID'].tolist()
```

```python
disease_tip_tipID = disease_tip['TIP_ID'].tolist()
```

```python
tipID_difference = set(disease_tip_tipID) - set(tip_tipID)
```

```python
tipID_difference
```

## Manual drop row

```python
drop_index = disease_tip[(disease_tip['TIP_ID'] == 100038) |(disease_tip['TIP_ID'] == 100039)].index
```

```python
drop_index
```

## Auto drop row

```python
# iterrows() to find all the TIP_ID (from tipID_difference) in disease_tip table and record their corresponding row index
# append the index to drop_index_list for the purpose of dropping them from disease_tip dataframe
drop_index_list = []
for idx,row in disease_tip.iterrows():
    for tip_id in tipID_difference:
        if row['TIP_ID'] == tip_id:
            drop_index_list.append(idx)
```

```python
drop_index_list
```

```python
disease_tip.drop(drop_index_list, inplace=True)
```

```python
disease_tip.to_csv('disease tip.csv', index=False)
```
