```python
import pandas as pd
```

```python
nsw12_14 = pd.read_excel('nsw 2012-14 air data.xls', sheet_name=0, skiprows=[0,1], header=0)
```

```python
nsw15_16 = pd.read_excel('nsw 2015-16 air data.xls', sheet_name=0, skiprows=[0,1], header=0)
```

```python
nsw17_18 = pd.read_excel('nsw site avg 2017-18 air data.xls', sheet_name=0, skiprows=[0,1], header=0)
```

```python
frames = [nsw12_14, nsw15_16, nsw17_18]
data = pd.concat(frames)
```

```python
data.head()
```

```python
dropna = data.dropna(axis=1, how='all')
```

```python
dropna
```

```python
names = dropna.columns.values
```

```python
pm10 = [x for x in names if 'PM10' in x]
```

```python
pm10_df = dropna[dropna.columns.intersection(pm10)]
```

```python
pm10_df_fillna = pm10_df.copy()
```

```python
pm10_df_fillna.fillna(pm10_df_fillna.mean(), inplace=True)
```

```python
pm10_df_fillna
```

```python
pm10_df_fillna['mean'] = pm10_df_fillna.mean(axis=1)
```

```python
pm10_df_fillna
```

```python
pm10_mean = pd.DataFrame()
```

```python
pm10_mean['Date'] = dropna['Date']
pm10_mean['PM10']=pm10_df_fillna.mean(axis=1)
```

```python
mean = pd.DataFrame()
mean['Date'] = dropna['Date']
mean['PM10']=pm10_df_fillna.mean(axis=1)
```

```python
no2 = [x for x in names if 'NO2' in x]
no2_df = dropna[dropna.columns.intersection(no2)]
no2_df_fillna = no2_df.copy()
no2_df_fillna.fillna(no2_df_fillna.mean(), inplace=True)
mean['NO2'] = no2_df_fillna.mean(axis=1)
```

```python
ozone = [x for x in names if 'OZONE' in x]
ozone_df = dropna[dropna.columns.intersection(ozone)]
ozone_df_fillna = ozone_df.copy()
ozone_df_fillna.fillna(ozone_df_fillna.mean(), inplace=True)
mean['OZONE'] = ozone_df_fillna.mean(axis=1)
```

```python
mean.to_excel('nsw avg all particles 2012-2017.xlsx', sheet_name='NSW', index=False)
```
