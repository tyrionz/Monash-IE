```python
import pandas as pd
```

```python
data = pd.read_csv('palmerston,-australia-air-quality.csv')
```

```python
data
```

```python
data['date'] = pd.to_datetime(data['date'], yearfirst=True)
```

```python
data.sort_values(by='date',inplace=True)
```

```python
data.reset_index(inplace=True,drop=True)
```

```python
data
```

```python
import numpy as np
data.replace(r'^\s*$', np.nan, regex=True, inplace=True)
```

```python
data
```

```python
data[' pm25'] = pd.to_numeric(data[' pm25'])
data[' pm10'] = pd.to_numeric(data[' pm10'])
data[' o3'] = pd.to_numeric(data[' o3'])
data[' no2'] = pd.to_numeric(data[' no2'])
data[' so2'] = pd.to_numeric(data[' so2'])
data[' co'] = pd.to_numeric(data[' co'])
```

```python
data
```

```python
data[' pm25'].fillna(data[' pm25'].mode()[0], inplace=True)
```

```python
data[' pm10'].fillna(data[' pm10'].mode()[0], inplace=True)
data[' o3'].fillna(data[' o3'].mode()[0], inplace=True)
data[' no2'].fillna(data[' no2'].mode()[0], inplace=True)
data[' so2'].fillna(data[' so2'].mode()[0], inplace=True)
data[' co'].fillna(data[' co'].mode()[0], inplace=True)
```

```python
data
```

```python
data.to_excel('Palmerston wrangled.xlsx', index=False)
```
