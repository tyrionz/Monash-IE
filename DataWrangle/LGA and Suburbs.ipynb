```python
import pandas as pd
```

```python
data = pd.read_excel('LocalityFinder.xlsx', skiprows = [0], header = 1)
```

```python
data.head()
```

```python
data.columns
```

```python
selected_df = data[['Locality Name', 'Municipality\r\nName']]
```

```python
selected_df
```

```python
selected_df.columns = ['SUBURB', 'LGA']
```

```python
selected_df
```

```python
# remove either city council or shire council in the LGA column
def remove_last2words(string):
    lga = ''
    if len(string.split()) == 3:
        lga = string.split()[:-2][0]
    else:
        lga_list = [words for words in string.split()[:-2]]
        for words in lga_list:
            lga += words + ' '
            
    return lga.strip() # remove last white space
```

```python
selected_df['LGA'] = selected_df['LGA'].apply(remove_last2words)
```

```python
selected_df
```

```python
selected_df.drop_duplicates(keep='first', inplace=True)
```

```python
selected_df
```

```python
# generate locality_ID
locality_ID = []
counter = 1000000
for i in range(len(selected_df)):
    counter += 1
    locality_ID.append(counter)
```

```python
# create final dataframe
final_df = pd.DataFrame({'LOCALITY_ID' : locality_ID,
                        'SUBURB': selected_df['SUBURB'],
                        'LGA': selected_df['LGA'].apply(lambda x: x.upper())})
```

```python
final_df
```
