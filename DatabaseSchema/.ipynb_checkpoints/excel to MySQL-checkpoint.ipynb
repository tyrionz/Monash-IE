```python
import xlrd
import MySQLdb

# Open the workbook and define the worksheet
book = xlrd.open_workbook("Disease table.xls")
sheet = book.sheet_by_name("Disease")

# Establish a MySQL connection
database = MySQLdb.connect (host="nhas-nobabyhosp.cf4yzwjind5x.us-east-1.rds.amazonaws.com", user = "nhasAdmin", passwd = "nhasBaby", db = "nhas-nobabyhosp")

# Get the cursor, which is used to traverse the database, line by line
cursor = database.cursor()

# Create the INSERT INTO sql query
query = """INSERT INTO DISEASE (DISEASE_ID, DISEASE_NAME) VALUES (%s, %s)"""

# Create a For loop to iterate through each row in the XLS file, starting at row 2 to skip the headers
for r in range(1, sheet.nrows):
    DISEASE_ID = sheet.cell(r,).value
    DISEASE_NAME = sheet.cell(r,1).value

    # Assign values from each row
    values = (DISEASE_ID, DISEASE_NAME)

    # Execute sql Query
    cursor.execute(query, values)

# Close the cursor
cursor.close()

# Commit the transaction
database.commit()

# Close the database connection
database.close()

# Print results
print ""
print "All Done! Bye, for now."
print ""
columns = str(sheet.ncols)
rows = str(sheet.nrows)
print "I just imported " %2B columns %2B " columns and " %2B rows %2B " rows to MySQL!"
```
